```python
def is_date_in_range(target_date, start, end):
    diff_from_start = target_date - start
    diff_to_end = end - target_date
    return int(diff_from_start.days >= 0 and diff_to_end.days >= 0)

def turnover_rate(df, start, end):
  seps = sum(map(lambda y_dt: is_date_in_range(y_dt, start, end), df[df["Estado"]=="Inactiva"]["Fecha de inicio de vigencia"]))
  A_employees = sum(map(lambda y_dt: y_dt <= start, df["Fecha de inicio de vigencia"]))
  B_employees = sum(map(lambda y_dt: y_dt <= end, df["Fecha de inicio de vigencia"]))
  return A_employees / B_employees * 100

```
